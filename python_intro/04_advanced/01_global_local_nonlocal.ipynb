### Global 

```python
b = 9
def foo():
    print("inside foo", b)

foo()
print("main", b)
```

###  Local

```python
def bar():
    y = 7
    print("inside bar", y)

bar()
print("main", y) # expect error
```

### Pay attention!

```python
b = 9
def magic():
    y = 7
    print("inside magic", y, b)
    b = 6

print("main", b)
magic()
```

```python
b = 9
def magic_local():
    y = 7
    b = 6
    print("inside magic_local", y, b)

magic_local()
print("main", b)
```

### `global` keyword

```python
b = 9
def magic_global():
    global b # <----
    y = 7
    print("inside magic_local", y, b)
    b = 6

magic_global()
print("main", b)
```

###  free variables (aka nonlocal variables)

```python
def maker():
    so_far=[]
    def avg(new_value):
        so_far.append(new_value)
        return sum(so_far)/len(so_far)
    return avg 

av = maker()
print(av(9))
print(av(11))
print(av(40))
```

```python
def maker_right():
    _sum = 0
    _count = 0
    def avg(new_value):
        nonlocal _sum,_count
        _sum += new_value
        _count += 1
        return _sum/_count
    return avg

av = maker_right()
print(av(9))
print(av(11))
print(av(40))
```
